```python
from sklearn.base import BaseEstimator, ClassifierMixin

class EuclideanClassifier(BaseEstimator, ClassifierMixin):  
    """Classify samples based on the distance from the mean feature value"""

    def __init__(self):
        self.X_mean_ = None
        


    def fit(self, X, y):
        a,b=X.shape
        self.X_mean_ =np.zeros((10,b))               #idia douleia me prin mono pou ulopoeitai ws methodoi 
        num=np.zeros(10)
        for d in range(0,10):                          #ths klashs euclideanclassifier
            for k in range(0,len(y)):
                if (y[k]==d):
                    num[d]+=1
                    for i in range(0,b):
                        self.X_mean_[d,i]+=X[k,i]
            self.X_mean_[d]/=num[d]
        return self
    


    def predict(self, X):
        """
        Make predictions for X based on the
        euclidean distance from self.X_mean_
        """
        a,_=X.shape
        pred=np.zeros(a)
        for i in range(0,len(X)):
            min=linalg.norm(self.X_mean_[0]-X[i,:])
            for k in range(0,10):
                if (linalg.norm(self.X_mean_[k,:]-X[i,:])<min):
                    min=linalg.norm(self.X_mean_[k,:]-X[i,:])
                    pred[i]=k
        return pred
        
        
    
    def score(self, X, y):
        """
        Return accuracy score on the predictions
        for X based on ground truth y
        """
        correct=0
        c=self.predict(X)
        for i in range(0,len(y)):
            if (c[i]==y[i]):
                correct+=1
        return correct/len(y)
        
```

Με λίγα λόγια, στην μέθοδο fit βρίσκουμε την μέση τιμή του κάθε pixel του κάθε ψηφίου,καθώς και την διασπορά του.Επίσης,όπως είπαμε και πριν υπολογίζουμε και τις a-priori πιθανότητες για κάθε κατηγορία.Για να υλοποιήσουμε τον Naive Bayes ταξινομητή στηριχτήκαμε στον υπολογισμό της πιθανότητας

Αξίζει να σημειωθεί ότι εφαρμόζοντας τον παραπάνω τύπο ,εάν έχουμε διασπορά 0 έχουμε διαίρεση με μηδέν.Γι' αυτόν το λόγο στον κώδικά μας καλύπτουμε αυτή την περίπτωση δίνοντας στην διασπορά μια πολύ μικρή τιμή , κοντά στο μηδέν.

Ερώτημα 15:β)Τώρα μας ζητείται ο υπολογισμός του σκορ για το Βήμα 15 α). Ο υπολογισμός του score για την υλοποίησή μας γίνεται στην μέθοδο score.

```python
import math

def gaussian(x, mu, sig): #tha xrisimopoihsoume kanoniki katanomh gia tis times tou kathe pixel
    return np.exp(-np.power(x - mu, 2.) / (2 * sig))/math.sqrt(2*math.pi*sig)

class mybayes():
    def __init__(self):
        self.mean=None
        self.var=None
        self.P_C=None
    
    def fit(self,X,y):
        self.mean =np.zeros((10,len(X[1]))) 
        num=np.zeros(10)
        for d in range(0,10):                          #upologizoume tis meses times tou kathe pixel 
            for k in range(0,len(y)):                  #akribws opws sth proergasia
                if (y[k]==d):
                    num[d]+=1
                    for i in range(0,len(X[1])):
                        self.mean[d,i]+=X[k,i]
            self.mean[d]/=num[d]
            
            
           
        self.var=np.zeros((10,len(X[1])))
        for d in range(0,10):                        #upologizoume tis diaspores
            for k in range(0,len(y)):
                if (y[k]==d):
                    for i in range(0,len(X[1])):
                        self.var[d,i]+=(self.mean[d,i]-X[k,i])**2
            self.var[d]/=num[d]
        
        self.P_C=num/len(y)                      #upologizoume a priori pithanotites kai oloklhrwnetai to modelo
        return self
    
    
    def predict(self,X):
        P_xc=np.ones((len(X),10))
        P_cx=np.ones((len(X),10))
        for s in range(0,len(X)):
            for k in range(0,10):
                for i in range(0,len(X[1])):
                    if (self.var[k,i]!= 0):
                        P_xc[s,k]=P_xc[s,k]*gaussian(X[s,i],self.mean[k,i],self.var[k,i])
                    else:
                        P_xc[s,k]=P_xc[s,k]*gaussian(X[s,i],self.mean[k,i],0.00001) #se periptwsh pou exoume mideniki diaspora
                P_cx[s,k]=P_xc[s,k]*self.P_C[k]


        nbc=np.zeros(len(X))
        for i in range(0,len(X)):
            nbc[i]=np.argmax(P_cx[i]) #h problepsh ginetai mesw tis kathgorias pou megistopoiei thn P_cx
        return nbc
    
    def score(self,X,y):
        corr=0
        for i in range(0,len(y)):
            if (X[i]==y[i]):
                    corr+=1
        return corr/len(y)
        
        
        
        
        
```

Για τον σκοπό αυτό υλοποιήσαμε μια κλάση digitsDataset, τα αντικείμενα της οποίας παίρνουν ως όρισμα ένα path για κάποιο αρχείο με δεδομένα.Η κλάση έχει 3 attributes,ένα το οποίο δίνει το πλήθος των sample και 2 πίνακες, έναν για το x_data και έναν για το y_data αντίστοιχα. Οι πίνακες αυτοί προκύπτουν διαχωρίζοντας το αρχείο που δόθηκε ως είσοδος.Μέρος της κλάσης αυτής είναι μια μέθοδος __getitem__ η οποία μας επιτρέπει να χειριζόμαστε τους πίνακες x_data και y_data μέσω indexes. Αντικείμενο αυτής της κλάσης ,με όρισμα το train αρχείο που μας δόθηκε, χρησιμοποιούμε ως όρισμα στην συνάρτηση DataLoader() της PyTorch η οποία μας χωρίζει τα δεδομένα με batch_size=32 και τα ανακατεύει.Επιστρέφει το train_loader με το οποίο γίνεται η εκπαίδευση του νευρωνικού δικτύου.



```python
import torch
import numpy as np
from torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader
```

```python
class digitsDataset(Dataset):
    
    def __init__(self,path):
        yx = np.loadtxt(path,dtype=np.float32)
        self.len = yx.shape[0]
        self.x_data = torch.from_numpy(yx[:,1:])
        self.y_data = torch.from_numpy(yx[:,[0]])
        
    def __getitem__(self, index):
        return self.x_data[index],self.y_data[index]
    
    def __len__(self):
        return self.len
```

```python
mydataset=digitsDataset('/home/giannis/train.txt')
train_loader = DataLoader(dataset=mydataset,batch_size=32,shuffle=True,num_workers=2)
```

β) Σε αυτό το ερώτημα υλοποιήσαμε ένα νευρωνικό δίκτυο σε μια υποκλάση της nn.Module. Το νευρωνικό αυτό αποτελείται από 4 layers . To 1o layer έχει 256 εισόδους ,όσα και τα pixels της εικόνας. Τα ενδιάμεσα layers επιλέχθηκαν αυθαίρετα να έχουν 64 νευρώνες εισόδου και 64 νευρώνες εξόδου και το τελευταίο layer έχει 64 νευρώνες εισόδου και 10 νευρώνες εξόδου όσες είναι και οι κατηγορίες ταξινόμησης στο πρόβλημά μας. Η μέθοδος forward πραγματοποιεί το πέρασμα παραμέτρων από το ένα layer στο επόμενο μέσω της συνάρτησης relu.

γ) Για το στάδιο αυτό ορίσαμε άλλη μια κλάση με 2 μεθόδους fit και score.Η μέθοδος fit δέχεται ως όρισμα το train_loader,που αποτελείται από τα δεδομένα εκπαίδευσης και εκπαιδεύει το μοντέλο . Ως loss function χρησιμοποιήσαμε την CrossEntropyLoss() και ως βελτιστοποιητή τον Adam(). Για την εκπαίδευση πραγματοποιήσαμε 4 περάσματα του dataset τα οποία με δοκιμές αποδείχθηκαν ικανά να κάνουν το μοντέλο να γενικεύσει χωρίς να αρχίζει να κάνει overfitting στα δεδομένα .Η μέθοδος score παίρνει ως όρισμα το mydataset2 το οποίο είναι αντικείμενο της κλάσης digitsDataset και έχει μέσα τα test δεδομένα.

```python
from torch import nn
import torch.nn.functional as F
from torchvision import datasets, transforms
import torch.optim as optim
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.layer1 = nn.Linear(16*16, 64)
        self.layer2 = nn.Linear(64, 64)
        self.layer3 = nn.Linear(64, 64)
        self.layer4 = nn.Linear(64,10)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.relu(self.layer3(x))
        x = self.layer4(x)
        return F.log_softmax(x, dim=1)
    

net = Net()

class trainNtestNet():
    def __init__(self):
        self.loss=None
    
    def fit(self,train_loader):
        loss_function = nn.CrossEntropyLoss()
        optimizer = optim.Adam(net.parameters(), lr=0.001)
        for epoch in range(4): # 4 full passes over the data
            for data,target in train_loader:  # `data` is a batch of data
                X, y = data,target# X is the batch of features, y is the batch of targets.
                y=y.type(torch.long)
                net.zero_grad()  # sets gradients to 0 before loss calc.
                output = net(X.view(-1,256))# pass in the reshaped batch
                self.loss = loss_function(output,y.squeeze(1))  # calc and grab the loss value
                self.loss.backward()  # apply this loss backwards thru the network's parameters
                optimizer.step()  # attempt to optimize weights to account for loss/gradients
            print(self.loss) # print loss. We hope loss (a measure of wrong-ness) declines! 
        return self

    
    def score(self,mydataset2):
        correct = 0
        total = 0
        with torch.no_grad():
            for data in mydataset2:
                X, y = data
                output = net(X.view(-1,256))
                for idx, i in enumerate(output):
                    if torch.argmax(i) == y[idx].type(torch.long):
                        correct += 1
                    total += 1
        return(correct/total)
        
        
```

```python
mydataset2=digitsDataset('/home/giannis/test.txt')
eval1=trainNtestNet()
eval1.fit(train_loader)
print(eval1.score(mydataset2))
```
